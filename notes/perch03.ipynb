#### 다중 회귀

- 여러 개의 특성을 사용한 선형 회귀

- 특성 공학 : 기존의 특성을 사용해 새로운 특성을 뽑아내는 작업

#### 데이터 준비
- 판다스 : 데이터 분석 라이브러리
- 데이터프레임 : 판다스의 핵심 데이터 구조

```python
import pandas as pd
#df = pd.read_csv('http://bit.ly/perch_csv_data')
df = pd.read_csv('../datas/perch_full.csv')
perch_full = df.to_numpy()
print(perch_full)
```

```python
import numpy as np

perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 훈련 데이터와 테스트 데이터로 나누기

from sklearn.model_selection import train_test_split

train_input,test_input,train_target,test_target = train_test_split(perch_full,perch_weight,random_state=42)
```

#### 사이킷런의 변환기

- 특성을 만들거나 전처리 하기 위해 사이킷런에서 제공하는 다양한 클래스
- 모두 fit과 transform() 메서드를 제공

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias=False)
poly.fit([[2,3]])
poly.transform([[2,3]])
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)

test_poly = poly.transform(test_input)

print(train_poly)

print(poly.get_feature_names_out())
```

#### 다중 회귀 모델 훈련하기


```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train_poly,train_target)

print(lr.score(train_poly,train_target))

print(lr.score(test_poly,test_target))

# 이로서 과소적합 문제는 해결했다.
# 특성이 늘어나면 선형 회귀의 능력은 매우 강하다
```

#### 특성을 더 많이 추가하면?

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
train_poly.shape



# 55(배열의 열의 개수가 특성의 개수)
```

```python
lr.fit(train_poly,train_target)
print("훈련 세트 {}".format(lr.score(train_poly,train_target)))
print("테스트 세트 {}".format(lr.score(test_poly,test_target)))
```

- 특성이 개수를 늘리면 선형 모데를 아주 강력해지나, 과대적합되므로 테스트 세트에서는 형편없는 점수를 만든다
